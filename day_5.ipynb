 Rotate a list by n positions.


```python
#sortcut method
a=[1,2,3,4,5]
n=2
n=n%len(a)
rot=a[-n:]+a[:-n]
rot

```

```python
#2nd method
a=[1,2,3,4,5]
n=2
length=len(a)
n=n % length
rotate_list=[]
for i in range(length-n,length):
  rotate_list.append(a[i])
for i in range(length-n):
  rotate_list.append(a[i])
print(rotate_list)
```
